```python
def calculate_highest_profit(ticker, date):
    profits = []
    
    df = dill.load(open('Prices/' + ticker + '.pkd', 'rb'))
    df = df['close'].reset_index()
    
    df = df[df['date'] >= date]
    
    if df.empty:
        raise Exception("No data for " + date)
    else:
        initial_price = df['close'].iloc[0]
    
        for i in range(12):
            date = (datetime.datetime.strptime(date, '%Y-%m-%d') + datetime.timedelta(days = 30)).strftime('%Y-%m-%d')
            df = df[df['date'] >= date]
            if df.empty:
                raise Exception("No data for " + date)
            else:
                new_price = df['close'].iloc[0]
                profits.append((initial_price - new_price, i + 1))
        
        return max(profits)
```

```python
calculate_highest_profit('AZN', '2020-01-01')
```
